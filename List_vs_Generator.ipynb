```python
#Using List
numbers_list=[i for i in range(1,6)]
print("List Output:")
for num in numbers_list:
    print(num)

#Using Generator
def number_generator():
    for i in range(1,6):
        yield i

num_gen=number_generator()
print("Generator Output:")
for num in num_gen:
    print(num)
```

```python
# Difference between List and Generator

#1. Memory Usage:
#   - List stores all elements in memory at once.
#   - Generator produces values one at a time and uses less memory.

#2. Performance:
#   - List is faster when accessing elements multiple times.
#   - Generator is efficient for large datasets and one-time iteration.

#3. Syntax:
#   - List uses square brackets [].
#   - Generator uses yield keyword or parentheses ().

#4. Reusability:
#   - List can be reused multiple times.
#   - Generator gets exhausted after one complete iteration.
```
